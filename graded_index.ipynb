```python
import numpy as np
import matplotlib.pyplot as plt

# --- Styling: larger fonts for journal-ready figure ---
plt.rcParams.update({
    "font.size": 16,          # base font size
    "axes.titlesize": 18,     # title font
    "axes.labelsize": 18,     # axis labels
    "xtick.labelsize": 16,    # tick labels
    "ytick.labelsize": 16,
})

# Parameters
L = 1.0
Nx, Ny = 600, 300  # grid resolution over 2x1 domain

# Grid over the rectangle: (x, y) in [0,2] x [0,1]
x = np.linspace(0.0, 2.0, Nx)
y = np.linspace(0.0, 1.0, Ny)
X, Y = np.meshgrid(x, y)

# Piecewise graded-index definition:
# Region 1: 0 <= x <= 1.0  --> n_{g1}(x,y) = 5*sqrt(1 - 0.356*(x^2 + y^2)/L^2)
# Region 2: 1.0 < x <= 2.0 --> n_{g2}(x)   = 1.2 + 0.6*x/L
ng = np.empty_like(X, dtype=float)

mask1 = (X <= 1.0)
mask2 = ~mask1  # (X > 1.0)

ng[mask1] = 5.0 * np.sqrt(1.0 - 0.356 * (X[mask1]**2 + Y[mask1]**2) / (L**2))
ng[mask2] = 1.2 + 0.6 * X[mask2] / L

# Contour plot (keep previous colorbar range unless specified)
fig = plt.figure(figsize=(8.5, 4.8))
cs = plt.contourf(X, Y, ng, levels=40, vmin=1.56, vmax=5.0)
plt.xlabel("x")
plt.ylabel("y")

title = (r"$n_{g1}(x,y)=5\sqrt{1-0.356(x^2+y^2)/L^2}$ for $0\leq x\leq 1$,  "
         r"$n_{g2}(x)=1.2+0.6\,x/L$ for $1< x\leq 2$")
plt.title(title)

plt.xlim(0, 2)
plt.ylim(0, 1)
plt.tick_params(axis='both', which='both', length=6, width=1.2)

# Thicken axis spines
for spine in plt.gca().spines.values():
    spine.set_linewidth(1.2)

# Colorbar
cbar = plt.colorbar(cs)
cbar.set_label(r"Graded index $n_g$", fontsize=18)
cbar.ax.tick_params(labelsize=16, width=1.2, length=6)

plt.tight_layout()

# Save figure
output_path = "/mnt/data/graded_index_2x1_piecewise.png"
plt.savefig(output_path, dpi=300)
output_path
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.path import Path

# --- Styling: larger fonts for journal-ready figure ---
plt.rcParams.update({
    "font.size": 16,          # base font size
    "axes.titlesize": 18,     # title font
    "axes.labelsize": 18,     # axis labels
    "xtick.labelsize": 16,    # tick labels
    "ytick.labelsize": 16,
})

# Irregular quadrilateral domain vertices (in order)
verts = np.array([
    [0.0,   0.0],
    [2.2,   0.0],
    [1.5,   1.2],
    [0.75,  1.0]
])

# Build polygon path
poly_path = Path(verts, closed=True)

# Grid covering the polygon's bounding box
Nx, Ny = 900, 500  # resolution
xmin, xmax = verts[:,0].min(), verts[:,0].max()
ymin, ymax = verts[:,1].min(), verts[:,1].max()

x = np.linspace(xmin, xmax, Nx)
y = np.linspace(ymin, ymax, Ny)
X, Y = np.meshgrid(x, y)

# Determine which grid points lie inside the polygon
points = np.vstack([X.ravel(), Y.ravel()]).T
inside = poly_path.contains_points(points).reshape(Y.shape)

# Graded-index function: n_g(x,y) = 1 + 2(x + y) (defined everywhere)
ng = 1.0 + 2.0 * (X + Y)

# Mask values outside the polygon; compute vmin/vmax from interior only
ng_masked = np.ma.array(ng, mask=~inside)
vmin = ng[inside].min()
vmax = ng[inside].max()

# Plot with the full dynamic range over the entire domain interior
fig = plt.figure(figsize=(8.2, 5.4))
cs = plt.contourf(X, Y, ng_masked, levels=40, vmin=vmin, vmax=vmax)

# Overlay polygon boundary
closed_verts = np.vstack([verts, verts[0]])
plt.plot(closed_verts[:,0], closed_verts[:,1], linewidth=2)

# Labels and title
plt.xlabel("x")
plt.ylabel("y")
plt.title(r"$n_g(x,y)=1+2(x+y)$")

# Limits, aspect, ticks
plt.xlim(xmin, xmax)
plt.ylim(ymin, ymax)
plt.gca().set_aspect('equal', adjustable='box')
plt.tick_params(axis='both', which='both', length=6, width=1.2)

# Thicken spines
for spine in plt.gca().spines.values():
    spine.set_linewidth(1.2)

# Colorbar using full interior range
cbar = plt.colorbar(cs)
cbar.set_label(r"Refractive index $n_g(x,y)$", fontsize=18)
cbar.ax.tick_params(labelsize=16, width=1.2, length=6)

plt.tight_layout()

# Save figure
output_path = "/mnt/data/graded_index_irregular_polygon_fullrange.png"
plt.savefig(output_path, dpi=300)
output_path
```

```python
# Guassian distribution
# Contour plot of graded-index refractive index n(x,y) on a unit square [0,1]x[0,1]
# - Uses matplotlib (no seaborn), single plot, no manual colors set
# - You can switch 'model' to 'gaussian', 'parabolic_grin', or 'anisotropic_gaussian'
# - Parameters are grouped for easy editing

import numpy as np
import matplotlib.pyplot as plt

# -------------------- Configuration --------------------
Lx, Ly = 1.0, 1.0          # unit square domain
Nx, Ny = 800, 800          # grid resolution

model = "anisotropic_gaussian"         # options: 'gaussian', 'parabolic_grin', 'anisotropic_gaussian'

# Parameters for the models
params = {
    "gaussian": {
        "a": 1.46,         # baseline/peak index (typical glass ~1.46)
        "b1": 150.0,       # inverse length^2 along x
        "b2": 150.0,       # inverse length^2 along y
        "x1": 0.4,         # center x
        "y1": 0.7,         # center y
        "offset": 0.0      # optional uniform offset to shift baseline
    },
    "parabolic_grin": {
        # Classic graded-index fiber: n(r)  n0*sqrt(1 - 2 (r/a)^2) for r <= a
        # Here we implement a circular core of radius a centered at (0.5,0.5)
        "n0": 1.46,
        "Delta": 0.01,     # fractional index difference (typical ~1%)
        "a_core": 0.35,    # core radius (in unit domain)
        "x1": 0.5,
        "y1": 0.5,
        "n_clad": 1.45     # cladding index outside core
    },
    "anisotropic_gaussian": {
        # Elliptical/astigmatic Gaussian
        "a": 1.46,
        "b1": 300.0,       # tighter along x
        "b2": 50.0,        # broader along y
        "x1": 0.4,
        "y1": 0.7,
        "offset": 0.0
    }
}

# -------------------- Grid --------------------
x = np.linspace(0.0, Lx, Nx)
y = np.linspace(0.0, Ly, Ny)
X, Y = np.meshgrid(x, y, indexing="xy")

# -------------------- Refractive index models --------------------
def n_gaussian(X, Y, p):
    # n(x,y) = offset + a * exp(-b1 (x-x1)^2 - b2 (y-y1)^2)
    dx = X - p["x1"]
    dy = Y - p["y1"]
    return p["offset"] + p["a"] * np.exp(-p["b1"] * dx**2 - p["b2"] * dy**2)

def n_parabolic_grin(X, Y, p):
    # n(x,y) = n0 * sqrt(1 - 2(r/a)^2) for r<=a, else n_clad
    dx = X - p["x1"]
    dy = Y - p["y1"]
    r = np.sqrt(dx**2 + dy**2)
    n = np.where(
        r <= p["a_core"],
        p["n0"] * np.sqrt(np.maximum(0.0, 1.0 - 2.0 * p["Delta"] * (r / p["a_core"])**2)),
        p["n_clad"]
    )
    return n

def n_anisotropic_gaussian(X, Y, p):
    dx = X - p["x1"]
    dy = Y - p["y1"]
    return p["offset"] + p["a"] * np.exp(-p["b1"] * dx**2 - p["b2"] * dy**2)

# -------------------- Evaluate model --------------------
if model == "gaussian":
    p = params["gaussian"]
    Nxy = n_gaussian(X, Y, p)
    title = f"$n_g(x,y) = {p['a']} exp[-{p['b1']}(x- {p['x1']}) - {p['b2']}(y-{p['y1']})]$"
elif model == "parabolic_grin":
    p = params["parabolic_grin"]
    Nxy = n_parabolic_grin(X, Y, p)
    title = f"$n_g(x,y) = {p['a']} exp[-{p['b1']}(x- {p['x1']}) - {p['b2']}(y-{p['y1']})]$"
elif model == "anisotropic_gaussian":
    p = params["anisotropic_gaussian"]
    Nxy = n_anisotropic_gaussian(X, Y, p)
    title = f"$n_g(x,y) = {p['a']} exp[-{p['b1']}(x- {p['x1']}) - {p['b2']}(y-{p['y1']})]$"
else:
    raise ValueError("Unknown model selected. Choose 'gaussian', 'parabolic_grin', or 'anisotropic_gaussian'.")

# -------------------- Plot --------------------
plt.figure(figsize=(6, 5), dpi=140)
# Use contourf without specifying colors
contours = plt.contourf(X, Y, Nxy, levels=30)
cbar = plt.colorbar(contours)
cbar.set_label("Graded Index $n_g(x,y)$")

plt.xlabel("x")
plt.ylabel("y")
plt.title(title)
plt.tight_layout()
```

```python
# Guassian distribution

# Corrected Python code for plotting graded-index refractive index on a unit square
# - Fixes typos ("laser" vs "leaser"), math signs, and LaTeX in the title
# - Uses matplotlib only, single plot, and no manual color settings
# - Three models: 'optical_fiber' (Gaussian), 'laser_beam' (broader Gaussian),
#   and 'hyperbolic' (saddle-type with opposite curvatures along x and y)
#
# You can copy this cell and edit parameters or switch `model` below.

import numpy as np
import matplotlib.pyplot as plt

# -------------------- Configuration --------------------
Lx, Ly = 1.0, 1.0          # unit square domain
Nx, Ny = 800, 800          # grid resolution

model = "optical_fiber"       # options: 'optical_fiber', 'laser_beam', 'hyperbolic'

# Parameters for the models
params = {
    "optical_fiber": {
        "a": 1.50,         # peak index near center
        "b1": 1000.0,      # inverse length^2 along x
        "b2": 1000.0,      # inverse length^2 along y
        "x1": 0.4,         # center x
        "y1": 0.7,         # center y
        "offset": 0.0      # optional uniform offset
    },
    "laser_beam": {
        "a": 1.00,
        "b1": 100.0,
        "b2": 100.0,
        "x1": 0.5,
        "y1": 0.5,
        "offset": 0.0
    },
    "hyperbolic": {
        "a": 1.00,
        "b1": 200.0,       # curvature along x
        "b2": 200.0,       # curvature along y
        "x1": 0.4,
        "y1": 0.7,
        "offset": 0.0
    }
}

# -------------------- Grid --------------------
x = np.linspace(0.0, Lx, Nx)
y = np.linspace(0.0, Ly, Ny)
X, Y = np.meshgrid(x, y, indexing="xy")

# -------------------- Refractive index models --------------------
def n_optical_fiber(X, Y, p):
    # Elliptical Gaussian core: n = offset + a * exp(-(b1 dx^2 + b2 dy^2))
    dx = X - p["x1"]
    dy = Y - p["y1"]
    return p["offset"] + p["a"] * np.exp(-(p["b1"] * dx**2 + p["b2"] * dy**2))

def n_laser_beam(X, Y, p):
    # Broader Gaussian profile
    dx = X - p["x1"]
    dy = Y - p["y1"]
    return p["offset"] + p["a"] * np.exp(-(p["b1"] * dx**2 + p["b2"] * dy**2))

def n_hyperbolic(X, Y, p):
    # Saddle-type: n = offset + a * exp(-b1 dx^2 + b2 dy^2)
    dx = X - p["x1"]
    dy = Y - p["y1"]
    return p["offset"] + p["a"] * np.exp(-p["b1"] * dx**2 + p["b2"] * dy**2)

# -------------------- Evaluate model --------------------
if model == "optical_fiber":
    p = params["optical_fiber"]
    Nxy = n_optical_fiber(X, Y, p)
    title = f"$n_g(x,y) = {p['a']} exp[-{p['b1']}(x- {p['x1']}) - {p['b2']}(y-{p['y1']})]$"
elif model == "laser_beam":
    p = params["laser_beam"]
    Nxy = n_laser_beam(X, Y, p)
    title = f"$n_g(x,y) = {p['a']} exp[-{p['b1']}(x- {p['x1']}) - {p['b2']}(y-{p['y1']})]$"
elif model == "hyperbolic":
    p = params["hyperbolic"]
    Nxy = n_hyperbolic(X, Y, p)
    title = f"$n_g(x,y) = {p['a']} exp[-{p['b1']}(x- {p['x1']}) - {p['b2']}(y-{p['y1']})]$"
else:
    raise ValueError("Unknown model selected. Choose 'optical_fiber', 'laser_beam', or 'hyperbolic'.")

# -------------------- Plot --------------------
plt.figure(figsize=(6, 5), dpi=140)
contours = plt.contourf(X, Y, Nxy, levels=30)
cbar = plt.colorbar(contours)
cbar.set_label("Graded Index $n_g(x,y)$")

plt.xlabel("x")
plt.ylabel("y")
plt.title(title)
plt.tight_layout()
```
